```python
import pandas as pd 
import matplotlib.pyplot as plt

import plotly.plotly as py 
from plotly import graph_objs as go
```

```python
file_path = '../data/18109003_excelRawData.csv'

df = pd.read_csv(file_path, index_col=0)
df.head()

start_column = 0
cutoff_column = 150

#defined start and cutoff_columns to tell program to only measure peak of first EPSP
#only want to measure first EPSP bc second was to experiment with channels' kinetics

df = df[start_column:cutoff_column]
df.head()
```

```python
#min_values = pd.DataFrame({'min': df.min()}).reset_index() 


min_values = pd.DataFrame({'sweep': df.columns.values, 'min': df.min() * -1})
min_values['sweep'] = pd.to_numeric(min_values['sweep'])

min_values.head()
```

```python
plt.plot(min_values['sweep'], min_values['min'])
```

```python
data = [go.Scatter(x=min_values['sweep'], y=min_values['min'], mode='lines+markers')]
layout = go.Layout(title='Sweep Plot', xaxis=dict(title='Sweep'), yaxis=dict(title='mV'))
fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```
